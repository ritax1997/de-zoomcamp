```python
# # Q3

query = """
SELECT 
    SUM(CASE WHEN trip_distance <= 1 THEN 1 ELSE 0 END) AS up_to_1_mile,
    SUM(CASE WHEN trip_distance > 1 AND trip_distance <= 3 THEN 1 ELSE 0 END) AS between_1_and_3_miles,
    SUM(CASE WHEN trip_distance > 3 AND trip_distance <= 7 THEN 1 ELSE 0 END) AS between_3_and_7_miles,
    SUM(CASE WHEN trip_distance > 7 AND trip_distance <= 10 THEN 1 ELSE 0 END) AS between_7_and_10_miles,
    SUM(CASE WHEN trip_distance > 10 THEN 1 ELSE 0 END) AS over_10_miles
FROM green_trip_data
WHERE LEFT(lpep_pickup_datetime::TEXT, 7) = '2019-10'
  AND LEFT(lpep_dropoff_datetime::TEXT, 7) = '2019-10';

"""

pd.read_sql(query, con=engine)
```

```python
## Q4
query = """
SELECT 
    DATE(lpep_pickup_datetime) AS day, 
    MAX(green_trip_data.trip_distance) AS trip_distance
FROM 
    green_trip_data
GROUP BY 
    DATE(lpep_pickup_datetime)
ORDER BY 
    trip_distance DESC
LIMIT 1;
"""

pd.read_sql(query, con=engine)
```

```python
# Question 5 Three biggest pickup zones PULocationID LocationID

three_biggest_pickup ="""
SELECT z."Zone", 
       SUM(g."total_amount") AS total_amount, 
       COUNT(*) AS trip_count
FROM green_trip_data g
LEFT JOIN zones z ON g."PULocationID" = z."LocationID"
WHERE DATE(g."lpep_pickup_datetime") = '2019-10-18'
GROUP BY z."Zone"
HAVING SUM(g."total_amount") > 13000
ORDER BY total_amount DESC
LIMIT 5;
"""
# WHERE DATE(lpep_pickup_datetime) = '2019-10-18'
pd.read_sql(three_biggest_pickup, con=engine)
```

```python
## Question 6. Largest tip tip_amount , zone "East Harlem North"
largest_tip ="""
SELECT z."Zone", 
       RANK() OVER (ORDER BY SUM(g."tip_amount") DESC) AS rk
FROM green_trip_data g
LEFT JOIN zones z ON g."PULocationID" = z."LocationID"
WHERE EXTRACT(YEAR FROM g."lpep_pickup_datetime") = 2019 
  AND z."Zone" = 'East Harlem North'
GROUP BY z."Zone"
LIMIT 5;
"""
# WHERE DATE(lpep_pickup_datetime) = '2019-10-18'
pd.read_sql(largest_tip, con=engine)
```
